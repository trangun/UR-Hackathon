```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import LeaveOneOut
from sklearn.metrics import mean_squared_error
import numpy as np



for i in range(2,100):
    pca_data = PCA(n_components = i)
    
    columns_name = []
    for j in range(1, i+1):
        name = 'pc '+ str(j)
        columns_name.append(name)
        
    principalComponents = pca_data.fit_transform(normalized_data.iloc[:,:-1])
    
    principal_data_Df = pd.DataFrame(data = principalComponents
             , columns = columns_name)
    principal_data_Df['y'] = y
    #print(principal_data_Df.head())
    print('number: ', i)
    if sum(pca_data.explained_variance_ratio_)>0.9:
        print('Explained variation per principal component: {}'.format(pca_data.explained_variance_ratio_))
        break
 
```

```python
pca_data = PCA(n_components = 137)
principalComponents = pca_data.fit_transform(normalized_data.iloc[:,:-1])
```

```python
columns_name = []
for j in range(1, 138):
    name = 'pc '+ str(j)
    columns_name.append(name)
print(columns_name)

```

```python
principal_data_Df = pd.DataFrame(data = principalComponents
             , columns = columns_name)
principal_data_Df['y'] = y
```

```python
principal_data_Df.head()
```

```python
y = principal_data_Df['y']
X = principal_data_Df[columns_name]
    
loo = LeaveOneOut()
loo.get_n_splits(X)


y_pred = []
for train_index, test_index in loo.split(X):
    #print("TRAIN:", train_index, "TEST:", test_index)
    X_train = X.iloc[train_index]
    X_test = X.iloc[test_index]
    y_train = y.iloc[train_index]
    y_test = y.iloc[test_index]
    #print(y_test)
    
    
    regr = RandomForestRegressor(max_depth=2, random_state=0)
    regr.fit(X_train, y_train)

    #y_pred = regr.predict(X_test)
    y_pred.append(regr.predict(X_test)*std + avg)
    
    #print(test_index, y_pred, y_before.iloc[test_index])
    #mse.append(mean_squared_error(y_test, y_pred))
print(y_pred)
#print(sum(mse)/len(mse))
#print('num of pc: ', i, ' mse: ', sum(mse)/len(mse) )
```

```python
# y_pred is the original prediction value 
mse = []
for i in range(80):
    each = y_pred[i]
    #print(each)
    mse.append(mean_squared_error(y_before.iloc[test_index], each))
print(len(mse))
print(sum(mse)/len(mse))
```
